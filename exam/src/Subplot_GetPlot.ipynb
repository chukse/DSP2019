```python
cell_count_Y = [0,0,0,0,0,0,0,0]

for i in range(len(time)-1):
    result = 0
    for j in Data[:,:,:,i]:
        
        for k in j:
            for l in k:
                result += l
                
                cell_count_Y[i+1] = result



plt.figure(figsize=(10,10))
plt.plot(time,cell_count_Y)
plt.errorbar(time, cell_count_Y, yerr=.05,fmt='o', color='black', ecolor='lightgray', linestyle='dotted')
plt.xlabel('Time[days]',fontsize=11)
plt.ylabel('Tumor Cell Count',fontsize=11)
plt.title('Gompertzian Fit to Rat\'s Brain Tumor Growth',fontsize=18)
plt.legend('Legend')
```
